# Figure 4: positron spectra  in 1D and 2D simulation setups

```python
import sys
sys.path.append('../scripts/')
import numpy as np
import matplotlib.pyplot as plt
import cross_section, read_files
import h5py
from scipy.constants import c, e, m_e, epsilon_0, mu_0, N_A, k, h, hbar, alpha
```

```python
# Normalisation factor for weight
omega_0 = 1.88e15
n_0 = m_e * epsilon_0 * omega_0**2 / e**2
norm_weight_2d = n_0 * ( c / omega_0 )**2
norm_weight_1d = n_0 * ( c / omega_0 )
norm_length = n_0 * ( c / omega_0 )
norm_ene = 1.e-6 * m_e * c**2 / e

# Usefull function to calculate integrals
def integral(datax, dataf):
    result = 0.
    L = len(dataf)
    for i in range(L-1):
        result += dataf[i] * (datax[i+1] - datax[i])
    return result

def arraycenter(x):
    """
        returns centered array
    """
    return np.array([(x[i]+x[i+1])/2 for i in range(len(x)-1)])
```

# 1D simulations

```python
# Physical parameters
Z = np.array([29., 29.])
k = 2000.0 # photon energy
sigma = np.array([cross_section.bh_cs(elem, k) for elem in Z])
cell_vol = np.array([8./2000., 8./2000.])
Vg = np.array([8.0, 8.0])

# Inputs parameters
ng = 1.e-4 # photon density
ni = 100. # ion density
dt = 0.002 # simulation timestep
N_dt = 1 # number of timesteps

min_p = 1.0
max_p = k - 1.

#--------------------------------------------------------------------
# Simulated spectrum
#--------------------------------------------------------------------

# Read the RAW file
folder = "lumi/1d/"
case = ["TCS_of_iCDF_of_Vec_on", "TCS_on_iCDF_on_Vec_on"]
file = '/MS/RAW/positrons/RAW-positrons-000001.h5'

# Array initialisation
N_case = len(case)
dNdp_raw_1d, dNdp_the_1d, gamp_raw_1d = [[] for n1 in range(N_case)], [[] for n1 in range(N_case)], [[] for n1 in range(N_case)]

for i_case in range(N_case):

    path = folder + case[i_case] + file
    quants = ['q', 'ene']
    data_raw = read_files.read_raw(path, quants)
    data_raw['gamma'] = data_raw['ene'] + 1.0

    # Total number of photons in the simulation
    number_raw = np.sum(data_raw['q']) * cell_vol[i_case]

    # Simulated spectrum for Bremsstrahlung
    N = 100
    dNdp_raw_1d[i_case], bin_edges = np.histogram(data_raw['gamma'], weights=data_raw['q'], bins=N, range=(min_p, max_p))
    dNdp_raw_1d[i_case] = np.array(dNdp_raw_1d[i_case], dtype='float64')

    # We arrange the photon energy axis
    gamp_raw_1d[i_case] = (bin_edges[1::] + bin_edges[0:-1]) / 2.

    # We normalise its integral by the number of photons in the simulation
    norm = integral(gamp_raw_1d[i_case], dNdp_raw_1d[i_case])
    dNdp_raw_1d[i_case] *= (number_raw/norm)

    # We convert it from plasma units to physical units
    dNdp_raw_1d[i_case] *= norm_weight_1d

    #--------------------------------------------------------------------
    # Theoretical spectrum
    #--------------------------------------------------------------------

    # Number of photons
    N_phot = ng * Vg[i_case]

    # Probability for pair creation
    var = sigma[i_case] * norm_length * ni * dt
    proba = 1.0 - np.exp(-var)

    # Total number of photons created
    number_the = N_phot * proba * N_dt

    # Theoretical spectrum for Bremsstrahlung
    dNdp_the_1d[i_case] = np.array([cross_section.bh_cs_dif(elem, k, Z[i_case]) for elem in gamp_raw_1d[i_case]])

    # We normalise its integral to the number of photons expected
    norm = integral(gamp_raw_1d[i_case], dNdp_the_1d[i_case])
    dNdp_the_1d[i_case] *= (number_the/norm)

    # We convert it from plasma units to physical units
    dNdp_the_1d[i_case] *= norm_weight_1d

    # Sanity check
    integral_raw = integral(gamp_raw_1d[i_case], dNdp_raw_1d[i_case])
    integral_theo = integral(gamp_raw_1d[i_case], dNdp_the_1d[i_case])
    rel_diff = 100 * abs(integral_theo - integral_raw) / integral_theo
    print('\nnumber of positrons = {:.2e}'.format(integral_raw))
    print('\nrel diff = {:.2f} %'.format(rel_diff))
    
```

# 2D simulations

```python
# --------------------------------------------------------------------
# Simulated spectrum
#--------------------------------------------------------------------

cell_vol = [0.05 * 0.05, 0.05 * 0.05]

# Read the RAW file
folder = "lumi/2d/"
case = ["TCS_of_iCDF_of_Vec_on", "TCS_on_iCDF_on_Vec_on"]
file = '/MS/RAW/positrons/RAW-positrons-000010.h5'

# Array initialisation
N_case = len(case)
dNdp_raw_2d, gamp_raw_2d = [[] for n1 in range(N_case)], [[] for n1 in range(N_case)]
integral_raw_2d = np.zeros(N_case)

for i_case in range(N_case):

    path = folder + case[i_case] + file
    quants = ['q', 'ene']
    data_raw = read_files.read_raw(path, quants)
    data_raw['gamma'] = data_raw['ene'] + 1.0

    # Total number of photons in the simulation
    number_raw = np.sum(data_raw['q']) * cell_vol[i_case]

    # Simulated spectrum for Bremsstrahlung
    N = 100
    dNdp_raw_2d[i_case], bin_edges = np.histogram(data_raw['gamma'], weights=data_raw['q'], bins=N, range=(1., 1.1*max(data_raw['gamma'])))
    dNdp_raw_2d[i_case] = np.array(dNdp_raw_2d[i_case], dtype='float64')

    # We arrange the photon energy axis
    gamp_raw_2d[i_case] = (bin_edges[1::] + bin_edges[0:-1]) / 2.

    # We normalise its integral by the number of photons in the simulation
    norm = integral(gamp_raw_2d[i_case], dNdp_raw_2d[i_case])
    dNdp_raw_2d[i_case] *= (number_raw/norm)

    # We convert it from plasma units to physical units
    dNdp_raw_2d[i_case] *= norm_weight_2d

    # Sanity check
    integral_raw_2d[i_case] = integral(gamp_raw_2d[i_case], dNdp_raw_2d[i_case])
    print('\nnumber of positrons = {:.2e}'.format(integral_raw_2d[i_case]))

rel_diff = 100 * abs(integral_raw_2d[1] - integral_raw_2d[0]) / integral_raw_2d[0]
print('\nrel diff = {:.2f} %'.format(rel_diff))
```
