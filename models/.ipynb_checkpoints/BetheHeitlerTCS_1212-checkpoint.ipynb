```python
def RunModelTCS(model, Z, k, minn = 26.173497595399795, maxx = 32.15150421307929):
    """
        run model with real, non-rescaled parameters
        min and max values of prepared data
    """
    # rescale parameters
    Z, k = np.log10(Z), np.log10(k)
    # model predict
    res = model.predict(np.array([[Z, k]]), verbose=False)[0,0]
    return 10**(-minn - (maxx-minn)*res )
```

```python
def ModelPredictTCS(model, datasetUnBalanced, minn = 26.173497595399795, maxx = 32.15150421307929):
    x = np.array(datasetUnBalanced)[:,0:-1]
    x[:,0] = np.log10(x[:,0])
    x[:,1] = np.log10(x[:,1])
    test_predUnb = model.predict( x )
    test_predUnb = test_predUnb.reshape(test_predUnb.shape[0])
    test_predUnb = 10**(-minn - (maxx-minn)*test_predUnb )
    return test_predUnb
```

```python
def plot_history_cdf(history):
    loss = history['loss']
    val_loss = history['val_loss']

    train_bin_cross = history['binary_crossentropy']
    val_bin_cross = history['val_binary_crossentropy']


    epochs = range(1, len(loss) + 1)
    fig, ax1 = plt.subplots()

    l1 = ax1.plot(epochs, loss, 'bo', label='Training loss mape')
    vl1 = ax1.plot(epochs, val_loss, 'b', label='Validation loss mape')
    ax1.set_title('Training and validation loss')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss (mape)')
    ax1.set_yscale('log')

    ax2 = ax1.twinx()
    ac2= ax2.plot(epochs, train_bin_cross, 'o', c="red", label='Training bin cross')
    vac2= ax2.plot(epochs, val_bin_cross, 'r', label='Validation bin cross')
    ax2.set_ylabel('bin cross')
    ax2.set_yscale('log')

    lns = l1 + vl1 + vac2 + ac2
    labs = [l.get_label() for l in lns]
    ax2.legend(lns, labs, loc="upper right")
    fig.tight_layout()
    fig.show()
```

```python
def saveFullModel(model, history):
    
    name = model._name
    model.save("models/{}.h5".format(name))
    
    # convert the history.history dict to a pandas DataFrame:     
    hist_df = pd.DataFrame(history.history) 
    
    # save to csv: 
    with open("models/{}.csv".format(name), mode='w') as f:
        hist_df.to_csv(f)
        
    #with open("models/{}.txt".format(name), mode='w') as f:
    #    f.write(str(minn)+' '+str(maxx))
```

```python
datasetBalanced = pd.read_csv('dataCPP/BH_cpp_tcs_9e6_bal3_250.csv', delimiter=',', header=None, index_col=None)
datasetBalanced.head()
```

```python
datasetBalanced = np.array(datasetBalanced)
# split
train_data, train_targets, val_data, val_targets, test_data, test_targets = train_test_split(datasetBalanced)
```

```python
# Build, fit model
batch_size = int(0.005 * train_data.shape[0])

model = models.Sequential()
model._name="BetheHeitlerTCS_1212"

model.add (BatchNormalization(input_dim = 2))

model.add (layers.Dense(12, activation="relu"))
model.add (layers.Dense(12, activation="relu"))

model.add (layers.Dense(1, activation="sigmoid"))
model.compile(optimizer = "adam", loss='mape', \
    metrics=['binary_crossentropy', 'kullback_leibler_divergence'])

model.summary()
print("model count_params =",model.count_params())
```

```python
history = model.fit(train_data, train_targets,
                    epochs = 300,
                    batch_size = batch_size,
                    validation_data = (val_data, val_targets),
                    verbose=True,
                    )
```

```python
saveFullModel(model, history)
```

```python
keys = model.metrics_names

# train set
values = model.evaluate(train_data, train_targets, verbose=0)
train_errors = dict(map(lambda i,j : (i,j) , keys,values))

# validation set
values = model.evaluate(val_data, val_targets, verbose=0)
val_errors = dict(map(lambda i,j : (i,j) , keys,values))

# test set
values = model.evaluate(test_data, test_targets, verbose=0)
test_errors = dict(map(lambda i,j : (i,j) , keys,values))
```

```python
print('train / val / test')
for key in model.metrics_names:
    print('{} = {:.2e}, {:.2e}, {:.2e}'.format(key, train_errors[key], val_errors[key], test_errors[key]))
```
