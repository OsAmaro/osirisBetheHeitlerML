```python
from scipy.optimize import fsolve
def bh_icdf(Z, k, u):
    gpk = fsolve(lambda gpk: bh_cdf(Z, k, gpk)-u, u)[0]
    return gpk
```

```python
def RunModeliCDF(model, Z, k, u, minn= -0.018343999999999916 , maxx= 0.02699499999999999):
    """
        run model with real, non-rescaled parameters
        min and max values of prepared data

        returns gp/k
    """
    # rescale parameters
    Z, k, u = np.log10(Z), np.log10(k), u
    # model predict
    res = model.predict(np.array([[Z, k, u]]), verbose=False)[0,0]
    return res*(maxx-minn)+minn + u
```

```python
def ModelPredictiCDF(model, datasetUnBalanced, minn= -0.018343999999999916 , maxx= 0.02699499999999999):
    x = np.array(datasetUnBalanced)[:,0:-1]
    x[:,0] = np.log10(x[:,0])
    x[:,1] = np.log10(x[:,1])
    test_predUnb = model.predict( x )
    test_predUnb = test_predUnb.reshape(test_predUnb.shape[0])
    test_predUnb = test_predUnb*(maxx-minn) + minn + np.array(datasetUnBalanced)[:,-2]
    return test_predUnb
```

```python
def plot_history_cdf(history):
    loss = history['loss']
    val_loss = history['val_loss']

    train_mape = history['mape']
    val_mape = history['val_mape']


    epochs = range(1, len(loss) + 1)
    fig, ax1 = plt.subplots()

    l1 = ax1.plot(epochs, loss, 'bo', label='Training loss mae')
    vl1 = ax1.plot(epochs, val_loss, 'b', label='Validation loss mae')
    ax1.set_title('Training and validation loss')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss (mae)')
    ax1.set_yscale('log')

    ax2 = ax1.twinx()
    ac2= ax2.plot(epochs, train_mape, 'o', c="red", label='Training mape')
    vac2= ax2.plot(epochs, val_mape, 'r', label='Validation mape')
    ax2.set_ylabel('mape')
    ax2.set_yscale('log')

    lns = l1 + vl1 + vac2 + ac2
    labs = [l.get_label() for l in lns]
    ax2.legend(lns, labs, loc="upper right")
    fig.tight_layout()
    fig.show()
```

```python
def saveFullModel(model, history):
    
    name = model._name
    model.save("models/{}.h5".format(name))
    
    # convert the history.history dict to a pandas DataFrame:     
    hist_df = pd.DataFrame(history.history) 
    
    # save to csv: 
    with open("models/{}.csv".format(name), mode='w') as f:
        hist_df.to_csv(f)
```

```python
datasetBalanced = np.array(datasetBalanced)
# split
train_data, train_targets, val_data, val_targets, test_data, test_targets = train_test_split(datasetBalanced)
```

```python
# Build, fit model
batch_size = int(0.005 * train_data.shape[0])

model = models.Sequential()
model._name="BetheHeitleriCDF051mae_8168"

model.add (BatchNormalization(input_dim = 3))

model.add (layers.Dense(8, activation="relu"))
model.add (layers.Dense(16, activation="relu"))
model.add (layers.Dense(8, activation="relu"))

model.add (layers.Dense(1, activation="sigmoid"))
model.compile(optimizer = "adam", loss='mae', \
    metrics=['binary_crossentropy', 'kullback_leibler_divergence', 'mape'])

model.summary()
print("model count_params =",model.count_params())
```

```python
history = model.fit(train_data, train_targets,
                    epochs = 800,
                    batch_size = batch_size,
                    validation_data = (val_data, val_targets),
                    verbose=True,
                    )
```

```python
saveFullModel(model, history)
```

```python
keys = model.metrics_names

# train set
values = model.evaluate(train_data, train_targets, verbose=0)
train_errors = dict(map(lambda i,j : (i,j) , keys,values))

# validation set
values = model.evaluate(val_data, val_targets, verbose=0)
val_errors = dict(map(lambda i,j : (i,j) , keys,values))

# test set
values = model.evaluate(test_data, test_targets, verbose=0)
test_errors = dict(map(lambda i,j : (i,j) , keys,values))
```

```python
print('train / val / test')
for key in model.metrics_names:
    print('{} = {:.2e}, {:.2e}, {:.2e}'.format(key, train_errors[key], val_errors[key], test_errors[key]))
```
